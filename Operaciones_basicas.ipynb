```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'Nombre': ['Ana', 'Juan', 'María', 'Carlos', 'Sofia'],
    'Edad' : [25, 30, 22, 28, 35],
    'Ciudad' : ['Madrid', 'Barcelona', 'Madris', 'Valencia', 'Barcelona'],
    'Puntuacion' : [85, 92, 78, 95, 88]
})

print("Dataframe Original: ")
print(df)
```

Filtrado

```python
mayores_25 = df[df['Edad'] >25]
print("\nPersonas mayores de 25: ")
print(mayores_25)
```

Agrupacion y agregacion

```python
resultado = df[df['Edad'] >25].groupby('Ciudad')['Puntuacion'].max()
print("\nPromedio de puntuacion por ciudad: ")
print(resultado)
```
