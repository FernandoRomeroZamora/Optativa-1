Llamar librerias

```python
import pandas as pd
import numpy as np
```

Crear una serie

```python
s = pd.Series([1, 3, 5, np.nan, 6, 8])
print("Serie: ")
print(s)
```

Explorar DataFrame

```python
df1 = pd.DataFrame(np.random.randn(6, 4), columns=list('ABCD'))
print("DataFrame: ")
print(df1)
df = pd.DataFrame({
    'A': [1,2,3,4],
    'B': pd.Timestamp('20130102'),
    'C': pd.Series(1, index=list(range(4)), dtype='float32'),
    'D': np.array([3] * 4, dtype='int32'),
    'E': pd.Categorical(["test", "train", "test", "train"]),
    'F': 'foo'
})
print("\nDataFrame: ")
print(df)
```

```python
print("\Tipos de datos: ")
print(df.dtypes)
print("\nPrimeras 2 filas: ")
print(df. head(2))
print("\nUltimas 2 filas: ")
print(df.tail(2))
```
